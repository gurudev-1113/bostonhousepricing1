```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)

data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
boston_df = pd.DataFrame(data, columns=feature_names)
boston_df['MEDV'] = target

display(boston_df.head())
```

INDEPENDENT AND DEPENDENT FEATURES

```python
y = dataset['Price']
x = dataset.drop(['MEDV', 'Price'], axis=1)
```

```python
y
```

```python
x
```

```python
#TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
X_train
```

MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

PRINT COEFFICIENTS

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
## PREDICTION WITH TEST DATA
```

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

ASSUMPTION


```python
## plot Scatter plot for prediction
```

```python
print(f"Length of y_test: {len(y_test)}")
print(f"Length of reg_pred: {len(reg_pred)}")
plt.scatter(y_test,reg_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print('Mean Absolute Error (MAE):', mean_absolute_error(y_test, reg_pred))
print('Mean Squared Error (MSE):', mean_squared_error(y_test, reg_pred))
print('R-squared (R2) score:', r2_score(y_test, reg_pred))
```
